Xi > 3:  
да: 50, 55, 61, 70  
Ŷ1 = 59  
нет: 100, 102, 103

Q = (50-59)^2 + (55-59)^2 + (61-59)^2 + (70-59)^2 + (100-101.7)^2 + (102-101.7)^2 + (103-101.7)^2 = 232.67

Разобьем первый узел:  

Xi > 6:  
да: 70  
Ŷ1 = 70  
нет: 50, 55, 61  
Ŷ2 ≈ 55.3  

Q = 0 + 5.3^2 + 0.3^2 + 5.7^2 = 60.67  

Попробуем другое разбиение:  

Xi > 5  
да: 61, 70  
Ŷ1 = 65.5  
нет: 50, 55  
Ŷ2 = 52.5  

Q = (4.5^2) * 2 + (2.5^2) * 2 = 53  

Второе разбиение оказалось лучше, другие разбиения будут хуже.  

Разобьем второй узел:  

Xi > 1:

нет: 100  
Ŷ2 = 100  

Q = (0.5^2) * 2 = 0.5  

Очевидно, другое разбиение будет хуже.
